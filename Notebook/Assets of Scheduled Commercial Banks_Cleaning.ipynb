```python
df.columns = df.columns.str.strip()
print(df.columns.tolist())
```

```python
df.drop([
'Investments outside India',
'Investments on government securities outside India',
'Investments on subsidiaries and/or joint ventures outside India',
'Other investments outside India',
'Advances outside India',
'Net inter-office adjustments by other assets',
'Investments on approved securities in India',
'Investments on subsidiaries and/or joint ventures in India',],axis=1,inplace=True)
```

```python
df['Cash in hand'] =  df['Cash in hand'].fillna(df['Cash in hand'].mean())
```

```python
df['Money at call and short notice'] = df['Money at call and short notice'].fillna(df['Money at call and short notice'].median())
```

```python
df['Other fixed assets'] =  df['Other fixed assets'].fillna(df['Other fixed assets'].mean())
```

```python


df['Interest accrued'] = df['Interest accrued'].fillna(
    df['Interest accrued'].mean()
)

df['Tax paid by other assets'] = df['Tax paid by other assets'].fillna(
    df['Tax paid by other assets'].mean()
)
```

```python
df['Stationery and Stamps by other assets'] = df['Stationery and Stamps by other assets'].fillna(
    df['Stationery and Stamps by other assets'].median()
)
```

```python
df['Others by other assets'] = df['Others by other assets'].fillna(
    df['Others by other assets'].mean()
)
```

```python
df['Balances with banks in India'] = df['Balances with banks in India'].fillna(
    df['Balances with banks in India'].mean()
)
```

```python
df['Balances with banks outside India'] = df['Balances with banks outside India'].fillna(df['Balances with banks outside India'].mean())

df['Investments'] = df['Investments'].fillna(df['Investments'].mean())

df['Investments in India'] = df['Investments in India'].fillna(df['Investments in India'].mean())

df['Investments on government securities in India'] = df['Investments on government securities in India'].fillna(df['Investments on government securities in India'].mean())
```

```python
# Fill missing values based on percentage

# Higher missing % → use median
df['Investments on shares in India'] = df['Investments on shares in India'].fillna(
    df['Investments on shares in India'].median()
)

df['Investments on debentures and bonds in India'] = df['Investments on debentures and bonds in India'].fillna(
    df['Investments on debentures and bonds in India'].median()
)

# Lower missing % → use mean
df['Other investments in India'] = df['Other investments in India'].fillna(
    df['Other investments in India'].mean()
)
```

```python
# Fill missing values with mean

df['Advances'] = df['Advances'].fillna(df['Advances'].mean())

df['Bills purchased and discounted'] = df['Bills purchased and discounted'].fillna(
    df['Bills purchased and discounted'].mean()
)

df['Cash credits, overdrafts and loans'] = df['Cash credits, overdrafts and loans'].fillna(
    df['Cash credits, overdrafts and loans'].mean()
)

df['Term loans'] = df['Term loans'].fillna(
    df['Term loans'].mean()
)

df['Advances secured by tangible assets'] = df['Advances secured by tangible assets'].fillna(
    df['Advances secured by tangible assets'].mean()
)

df['Advances covered by bank or government guarantees'] = df['Advances covered by bank or government guarantees'].fillna(
    df['Advances covered by bank or government guarantees'].mean()
)

df['Unsecured advances'] = df['Unsecured advances'].fillna(
    df['Unsecured advances'].mean()
)

df['Advances in India'] = df['Advances in India'].fillna(
    df['Advances in India'].mean()
)

df['Advances in priority sectors in India'] = df['Advances in priority sectors in India'].fillna(
    df['Advances in priority sectors in India'].mean()
)
```

```python
df['Advances in public sectors in India'] = df['Advances in public sectors in India'].fillna(
    df['Advances in public sectors in India'].median()
)

df['Advances in banks in India'] = df['Advances in banks in India'].fillna(
    df['Advances in banks in India'].median()
)
```

```python
df['Fixed assets under construction'] = df['Fixed assets under construction'].fillna(
    df['Fixed assets under construction'].median()
)
```

```python
# Fill missing values with mean

df['Other advances in India'] = df['Other advances in India'].fillna(
    df['Other advances in India'].mean()
)

df['Fixed assets'] = df['Fixed assets'].fillna(
    df['Fixed assets'].mean()
)

df['Premises by fixed assets'] = df['Premises by fixed assets'].fillna(
    df['Premises by fixed assets'].mean()
)
```
